```python

import numpy as np
from numpy import random
import matplotlib.pyplot as plt
from PIL import Image 
import PIL
import cv2

'''
methods for sampling data 
inputs : facies / num_samples / facies_data 
# method 1: take  sample from random positions  
# take  sample from random positions 

'''

def create_samples_from_facies(facies_data,facies_name,num_samples):
  height=len(facies_data)
  print(height)
  low = 1

  high = height-192
  print(high)
  sample_idx = random.randint(low, high, size=num_samples, dtype=int)
  # this should be converted into a new function / so the function should be classed insude the for loop 
  for item in sample_idx:
    image_name = str(item)+'.png'
    path_to_write = os.path.join('/content/drive/MyDrive/Projects/Roozbeh/DATASET',facies_name,image_name)
    # print(path_to_write)
    image_data = facies_data[item:item+192] 
    cv2.imwrite(path_to_write, np.array(image_data))
```
